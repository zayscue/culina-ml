```python
import pandas as pd
from ast import literal_eval
```

```python
df_raw_recipes = pd.read_csv('../data/RAW_recipes.csv')
```

```python
df_raw_recipes.head()
```

```python
df_recipes = pd.read_csv('../data/recipes.csv')
```

```python
df_recipes.head()
```

```python
df_recipe_popularity = pd.read_csv("../data/recipe_popularity.csv")
```

```python
df_recipe_popularity.head()
```

```python
df_recipes = pd.merge(
    df_recipes,
    df_recipe_popularity[["id", "ratings_weighted_average"]],
    on=["id"]
)
```

```python
df_recipe_content = pd.merge(
    df_recipes[["source_id", "id", "name", "ratings_weighted_average"]],
    df_raw_recipes[["id", "tags", "ingredients"]],
    left_on=["source_id"],
    right_on=["id"],
    how="left"
)
df_recipe_content = df_recipe_content.rename(columns={"id_x": "id"})[["id", "name", "tags", "ingredients", "ratings_weighted_average"]]
df_recipe_content["tags"] = df_recipe_content["tags"].apply(literal_eval).fillna("").astype("str")
df_recipe_content["ingredients"] = df_recipe_content["ingredients"].apply(literal_eval).fillna("").astype("str")
```

```python
df_recipe_content.head()
```

```python
df_top_10000_recipe_content = df_recipe_content.sort_values(["ratings_weighted_average"], ascending=False).reset_index(drop=True).head(10000)
```

```python
df_top_10000_recipe_content
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tf = TfidfVectorizer(analyzer="word", ngram_range=(1,2), min_df=0, stop_words="english")
tags_tfidf_matrix = tf.fit_transform(df_top_10000_recipe_content["tags"])
ingredients_tfidf_matrix = tf.fit_transform(df_top_10000_recipe_content["ingredients"])
```

```python
tags_tfidf_matrix.shape
```

```python
ingredients_tfidf_matrix.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
tags_cosine_sim = cosine_similarity(tags_tfidf_matrix,tags_tfidf_matrix)
ingredients_cosine_sim = cosine_similarity(ingredients_tfidf_matrix, ingredients_tfidf_matrix)
```

```python
tags_cosine_sim[:4, :4]
```

```python
ingredients_cosine_sim[:4, :4]
```

```python
ids = df_top_10000_recipe_content["id"]
indices = pd.Series(df_top_10000_recipe_content.index, index=df_top_10000_recipe_content["id"])
```

```python
ids
```

```python
indices
```

```python
df_recipe_tag_similarity = pd.DataFrame()
for idx in range(10000):
    tag_sim_scores = list(enumerate(tags_cosine_sim[idx]))
    tag_sim_scores = sorted(tag_sim_scores, key=lambda x: x[1], reverse=True)
    tag_sim_scores = tag_sim_scores[1:21]
    similar_recipes_by_tag = [{"recipe_id": ids[idx], "similar_recipe_id": ids[i[0]], "sim_score": i[1]} for i in tag_sim_scores]
    df_tmp = pd.DataFrame(similar_recipes_by_tag)
    df_recipe_tag_similarity = pd.concat([df_recipe_tag_similarity, df_tmp])
```

```python
df_recipe_ingredient_similarity = pd.DataFrame()
for idx in range(10000):
    ingredient_sim_scores = list(enumerate(ingredients_cosine_sim[idx]))
    ingredient_sim_scores = sorted(ingredient_sim_scores, key=lambda x: x[1], reverse=True)
    ingredient_sim_scores = ingredient_sim_scores[1:21]
    similar_recipes_by_ingredient = [{"recipe_id": ids[idx], "similar_recipe_id": ids[i[0]], "sim_score": i[1]} for i in ingredient_sim_scores]
    df_tmp = pd.DataFrame(similar_recipes_by_ingredient)
    df_recipe_ingredient_similarity = pd.concat([df_recipe_ingredient_similarity, df_tmp])
```

```python
df_recipe_tag_similarity.to_csv('../data/recipe_tag_similarity.csv', index=False)
df_recipe_ingredient_similarity.to_csv('../data/recipe_ingredient_similarity.csv', index=False)
```
